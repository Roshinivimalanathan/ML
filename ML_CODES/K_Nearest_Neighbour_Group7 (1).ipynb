```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
column_names = ["Name","Landmass","Zone","Area","Population","Language","Religion","Num_of_Colors","Main_Hue","Circles",
"Crosses","Saltires","Quarters","Sunstars","Crescent","Triangle","Icon","Animate","Text","Civic_Flag","Stripe_Colors","Red","Green","Blue","Gold","White","Black","Orange","Borders","Bars"
]
```

```python
data=pd.read_csv("flag.data",names=column_names)
```

```python
data
```

```python
X = data[["Circles", "Crosses"]]
Y = data[["Red", "Green"]]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=3)
```

```python
knn_classifier.fit(X_train, Y_train)
```

```python
predictions = knn_classifier.predict(X_test)
```

```python
sns.pairplot(data[["Circles", "Crosses", "Red", "Green", "Landmass"]], hue="Landmass")
plt.show()


plt.figure(figsize=(10, 8))
sns.heatmap(data[["Circles", "Crosses", "Red", "Green", "Landmass"]].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()


plt.figure(figsize=(8, 6))
sns.countplot(x="Landmass", data=data)
plt.title("Count Plot for Landmass")
plt.show()
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d

vor = Voronoi(data[["Circles", "Crosses"]])

fig = voronoi_plot_2d(vor, show_vertices=False, line_colors='orange', line_width=2, line_alpha=0.6, point_size=2)

plt.show()
```
